```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
from sklearn import linear_model
```

```python
raw_csv = pd.read_csv('life_data.csv')
income = {x['LOCATION']: x['Value'] for i,x in raw_csv.iterrows() if x['INDICATOR'] == 'IW_HADI'}
qual = {x['LOCATION']: x['Value'] for i,x in raw_csv.iterrows() if x['INDICATOR'] == 'SC_SNTWS'}
plt.plot(list(income.values()), list(qual.values()), "o")
plt.show()
```

```python
X = np.c_[list(income.values())]
y = np.c_[list(qual.values())]
```

```python
lin_model = linear_model.LinearRegression()
lin_model.fit(X, y)
```

```python
plt.plot(list(income.values()), list(qual.values()), "o")
plt.plot(X, lin_model.predict(X), color='blue')
for n in list(income.keys()):
    plt.annotate(n, (income[n], qual[n]))
plt.show()
```
